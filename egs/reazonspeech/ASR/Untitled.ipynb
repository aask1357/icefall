```python
import MeCab

mecab = MeCab.Tagger()
def encode(text):
    parsed_text = mecab.parse(text)
    katakana = ""

    for line in parsed_text.splitlines():
        if line == 'EOS':
            break
        parts = line.split('\t')
        katakana = f"{katakana}{parts[1]}"

    return katakana

print(encode("今も相手にロンバルドのほうに肩口で握られてもすぐさま流れを切る引き込み返しに変えたと。"))
print(encode("今日は天気が寒いです"))
```
